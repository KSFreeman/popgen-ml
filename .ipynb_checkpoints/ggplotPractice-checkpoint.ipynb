```python
midwestFile = "~/LOTTERHOS_LAB/machine_learning/midwest.csv"
midwest = pd.read_csv(midwestFile, sep = ",")

p = ggplot(midwest, aes(x="area", y="poptotal")) + geom_point() + geom_smooth(method = "lm")
p + xlim([0, 0.1]) + ylim([0,1000000])

g1 = p + coord_cartesian(xlim = [0,0.1], ylim=[0,1000000])
g1 +  labs (title="Area vs Population", subtitle="From midwest dataset", y="Population",
           x="Area", caption="Midwest Demographics")
g1 + ggtitle("Area Vs Population") + xlab("Area") + ylab("population")
```
